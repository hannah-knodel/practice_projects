# U.S. Medical Insurance Costs - Practice Exploratory Data Analysis

```python
# Objectives:
# Inspect dataset
# Find if there is a correlation between the increase in BMI and increase in charges
```

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
insurance_data = pd.read_csv('insurance.csv')
#print(insurance_data.head())

# Rename columns
insurance_data = insurance_data.rename(columns={'age': 'Age', 'sex': 'Sex', 'bmi': 'BMI', 'children': 'Children', 'smoker': 'Smoker', 'region': 'Region', 'charges': 'Charges'})
#print(insurance_data.head())

print(insurance_data.describe(include='all'))
print(insurance_data.dtypes)
```

```python
# Inspect data - this dataset has already been cleaned, but let's still run through the columns for null values too!
print(insurance_data.isnull().value_counts())

# BMI column inspection 
print(insurance_data.value_counts('BMI'))
# BMI exploration
average_BMI = insurance_data['BMI'].mean()
print('The average BMI for this dataset is: ' + str(average_BMI))
med_BMI = insurance_data['BMI'].median()
print('The median BMI for this dataset is: ' + str(med_BMI))
mode_BMI = insurance_data['BMI'].mode()[0]
print('The mode BMI for this dataset is: ' + str(mode_BMI))
BMI_spread = insurance_data['BMI'].max() - insurance_data['BMI'].min()
print('The BMI range for this dataset is: ' + str(BMI_spread))
BMI_std = insurance_data['BMI'].std()
print('The standard deviation in BMI for this dataset is: ' + str(BMI_std))
mad_BMI = (insurance_data['BMI'] - insurance_data['BMI'].mean()).abs().mean()
print('The mean absolute deviation in BMI is: ' + str(mad_BMI))
```

```python
# Plot histogram to see the fluctuation
sns.histplot(x='BMI', data=insurance_data)
plt.show()
plt.clf()
```

```python
# Boxplot for BMI
sns.boxplot(x='BMI', data=insurance_data)
plt.show()
plt.clf()
```

```python
# Scatter plot for BMI and Charges
plt.scatter(x=insurance_data['BMI'], y=insurance_data['Charges'])
plt.show()
plt.clf()
```

```python
# Covariance for BMI and charges
cov_BMI_charges = np.cov(insurance_data['BMI'], insurance_data['Charges'])
print('The covariance for BMI and insurance charges is: ' + str(cov_BMI_charges[1][0]))
# Correlation between BMI and charges
corr_BMI_charges, p = pearsonr(insurance_data['BMI'], insurance_data['Charges'])
print('The correlation between BMI and Charges is: ' + str(corr_BMI_charges))
```
